```python
import seaborn as sns
import ssl
ssl._create_default_https_context = ssl._create_unverified_context 
df=sns.load_dataset('tips') #titanic,iris,diamonds,penguin
df
```

```python
df.dtypes
```

```python
#feature selection==> required columns selection
correlation=df.corr(numeric_only=True)
correlation #a correlation matrix is formed , diagional elements are one
```

```python
#heatmap ==> visualise correlation matrix
sns.heatmap(correlation) # light color ==> strong correlation , dark ==> weak correlation
#cmap="coolwarm" ==> darker==> strong , light==> weak
```

```python
#jointplot ==> numerical values relation
sns.jointplot(x="total_bill",y="tip",data=df,kind="reg")
#right side==> histogram of tip
# top ==> histogram of total_bill
#scatter of both tip and total_bill
#kind="hex" ==> scatter is represented in hexagon shape
# kind="reg" ==> regression plot and distribution pattern of histogram  
```

```python
#pairplot==> numerical relation of multiple (correlation metrix)
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex") # hue ==> for specific column (categorical)
```

```python
#count plot ==> count of categorical colm
sns.countplot(x="sex",data=df)
```

```python
sns.countplot(x="smoker",data=df)
```

```python
sns.countplot(x="day",data=df,hue="smoker") # hue==> condition/parameter 
```

```python
#bar plot ==> comparison btw numerical and categorical data
sns.barplot(x="total_bill",y="day",data=df)
# black line ==> error bar(longer the bar ==> more variation) 
#ci=None ==>confidence interval (to remove error bar)
```

```python
sns.barplot(x="tip",y="day",data=df)
```

```python
#box plot ==>find outlier(visualise)
sns.boxplot(x="smoker",y="total_bill",data=df)
#whisker==> line extending from the box
#upper whisker==>maximum value
#lower whisker==> minimum value
#Q1==> 25th percentile(both line of box)
#Q2==> median (middle)
#Q3==> 75th percentile (upper line of box)
#iqr==> inter quartile range(blue box) 
#circles==> outliers
```
