```python
df.drop(["ID"],axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()
df["Employment Type"]=encode.fit_transform(df["Employment Type"])
df["GraduateOrNot"]=encode.fit_transform(df["GraduateOrNot"])
df["FrequentFlyer"]=encode.fit_transform(df["FrequentFlyer"])
df["EverTravelledAbroad"]=encode.fit_transform(df["EverTravelledAbroad"])
df
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import classification_report 
cr=classification_report(y_test,y_pred)
print(cr)
```
