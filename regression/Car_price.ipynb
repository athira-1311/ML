```python
#get_dummies
df1=pd.get_dummies(df[["Location","Fuel_Type","Transmission","Owner_Type","Brand"]],drop_first=True,dtype=int) #dtype==> false to numeric
df1
```

```python
df2=pd.concat([df,df1],axis=1)
df2
```

```python
df2.drop(["Unnamed: 0","Name","Location","Fuel_Type","Transmission","Owner_Type","New_Price","Brand"],inplace=True,axis=1)
df2
```

```python
df2["Mileage"]=df2["Mileage"].str.replace("km/kg","")
df2["Mileage"]=df2["Mileage"].str.replace("kmpl","")
df2["Engine"]=df2["Engine"].str.replace("CC","")
df2["Power"]=df2["Power"].str.replace("bhp","")

df2
```

```python
#mileage,engine,power has null values converting it into ==> 0(as object)
#null cant be treated as missing value since its object
df2["Mileage"]=df2["Mileage"].str.replace("null","0")
df2["Engine"]=df2["Engine"].str.replace("null","0")
df2["Power"]=df2["Power"].str.replace("null","0")
df2 
```

```python
df2.dtypes
```

```python
#converting datatype (power,engine,mileage)
df2["Mileage"]=df2["Mileage"].astype(float) #mileage colm has float values
df2["Engine"]=df2["Engine"].astype(float)
df2["Power"]=df2["Power"].astype(float)
df2
```

```python
df2.dtypes
```

```python
#power, mileage,engine cant have 0 as its values ,need to convert to missing value 
#null==> 0(object)==> dtype(float)==> missing value

df2.loc[df2.Mileage==0,"Mileage"]=np.nan  #nan==>missing value
df2.loc[df2.Power==0,"Power"]=np.nan
df2.loc[df2.Engine==0,"Engine"]=np.nan
```

```python
df2["Mileage"]=df2["Mileage"].fillna(df2["Mileage"].mean())
df2["Engine"]=df2["Engine"].fillna(df2["Engine"].mean())
df2["Power"]=df2["Power"].fillna(df2["Power"].mean())
df2["Seats"]=df2["Seats"].fillna(df2["Seats"].mode()[0])

```
