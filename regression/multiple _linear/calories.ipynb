```python
numeric_cols=df.select_dtypes(include=[np.number]).columns

plt.figure(figsize=(9,6))
sns.boxplot(data=df[numeric_cols])
plt.title("Boxplot of all features")
```

```python
outlier_columns=["Height","Weight","Heart_Rate","Body_Temp","Calories"]

def removal_outliers(df,columns):
    for i in columns:
        Q1=df[i].quantile(0.25)
        Q3=df[i].quantile(0.75)
        IQR=Q3-Q1

        lower_bound=Q1-1.5*IQR
        upper_bound=Q3+1.5*IQR
        df=df[(df[i]>=lower_bound)&(df[i]<=upper_bound)]
    return df
df_cleaned=removal_outliers(df,outlier_columns)
```

```python
correlation_matrix=df_cleaned.corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",fmt=".2f",linewidths=0.5) #annot=True ==>to provide values in cell(square box),
#fmt=".2f"==> 2 decimal value
plt.title("Heatmap")
```

```python
df_selected=df_cleaned.drop(columns=["Height","Gender"])

correlation_matrix=df_selected.corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",fmt=".2f",linewidths=0.5)

plt.title("Heatmap")
```

```python
x=df_selected.iloc[:,:-1]
y=df_selected.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
mape=mean_absolute_percentage_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
mape
```
