```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("customer1.txt",header=None)
df.columns=["id","fname","lname","age","prof","location"]
df
```

```python
df1=df["location"].value_counts()
df1
```

```python
plt.bar(df1.index,df1.values) #(x==> index i.e loccation , y==> value i.e count)
plt.xlabel("Location")
plt.ylabel("count")
plt.title("Each location count",fontsize=15)
```

```python
plt.pie(df1.values,labels=df1.index,autopct="%1.0f%%")
plt.legend(loc="upper left")
plt.tight_layout()
plt.title("Location count",fontsize=20)
```
