```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("company_sales_data.csv")
df
```

```python
#sales_dataset_matplotlib

# 1. Read Total profit of all months and show it using a line plot


# 2. Get total profit of all months and show line plot with the following Style properties
#  Generated line plot must include following Style properties:
#  Line Style dotted and Line-color should be red
#  Show legend at the lower right location.
#  X label name = Month Number
#  Y label name = Sold units number
#  Add a circle marker.
#  Line marker color as read
#  Line width should be 3

# 3. Read all product sales data and show it  using a multiline plot
#     square marker with color=green 

# 4.Read toothpaste sales data of each month and show it using a scatterplot

# 5.Read face cream and facewash product sales data and show it using the bar chart with grid line

# 6.Read the total profit of each month and show it using the histogram to see the most common profit ranges

# 7.Calculate total sale data for last year for each product and show it using a Pie chart

# 8. Read Bathing soap facewash of all months and display it using the Subplot
```

```python
#1
plt.plot(df["month_number"],df["total_profit"])
plt.xlabel("Month  Number")
plt.ylabel("Total Profit")
plt.title("Total Profit Over Months")
plt.xticks(df["month_number"]) # to define values according to x 
plt.yticks([100000,200000,300000,400000,500000,600000,700000])
```

```python
#2
plt.plot(df["month_number"],df["total_profit"],marker="o",linestyle="dotted",color="r",mec='r',mfc='r',linewidth=3,label="Profit Data")
plt.legend(loc="lower right")
plt.xlabel("Month  Number")
plt.ylabel("Sold Units Number")
plt.title("Styled Line Plot of Total Profit")
plt.xticks(df["month_number"]) # to define values according to x 
plt.yticks([100000,200000,300000,400000,500000,600000,700000])
plt.legend(loc="lower right")
```

```python
#3 
products=["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
for i in products:
    plt.plot(df["month_number"],df[i],marker="s",color="g")
plt.xlabel("Month Number")
plt.ylabel("Product sales")
plt.title("Multline Plot for all product sales")
plt.xticks(df["month_number"]) # to define values according to x 
plt.yticks([1000,2000,3000,4000,5000,6000,7000,8000])
```

```python
#4

plt.scatter(df["month_number"],df["toothpaste"])
plt.xlabel("Month Number")
plt.ylabel("Toothpaste Sales")
plt.title("Toothpaste Sales data of each month")
```

```python
#5
x=df["month_number"]
facecream=df["facecream"]
facewash=df["facewash"]
plt.bar([a-0.25 for a in x],facecream,width=0.25,align="edge",label="Face cream sales")
plt.bar([a+0.25 for a in x],facewash,width=-0.25,align="edge",label="Face wash sales")
plt.grid()
plt.xlabel("Month number")
plt.ylabel("Facecream and Facewash Sales")
plt.title("Face cream and face wash sales")
plt.legend(loc="upper right")
plt.tight_layout()
plt.xticks(x)
```

```python
#6
plt.hist(df["total_profit"])
plt.xlabel("Profit Range")
plt.ylabel("Frequency")
plt.title("Histogram of Total Profits")
```

```python
#7
total_sales=df[products].sum()
plt.pie(total_sales,autopct="%1.0f%%",labels=products)
plt.legend(loc="upper left")
plt.tight_layout()
plt.title("Total Sales data for last year for each product")
```

```python
#8
plt.subplot(1,2,1)#1st==> row count,2nd==>column 3rd==>position 
plt.plot(df["month_number"],df["bathingsoap"])
plt.xlabel("Month number")
plt.ylabel("Bathing soap sales units")
plt.title("Graph 1")


plt.subplot(1,2,2)
plt.plot(df["month_number"],df["facewash"])
plt.xlabel("Month number")
plt.ylabel("Facewash sales unit")
plt.title("Graph 2")
plt.suptitle("Bathingsoap and Facewash sales for all months")
plt.tight_layout()
```
